```python
from __future__ import division
import time
import networkx as nx
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.special import expit
```

```python
def process(u):
    return tanh(u)
    # return expit(u)
    # return np.minimum(np.maximum(u, -1), 1)

def external_in(N_in, t):
    A = 0.5
    omega = rnd.rand(N_in, 1)*10
    phi = rnd.rand(N_in, 1)
    _t = omega*t/(2*np.pi)
    return A*(np.sin(_t + phi))
```

```python
N_units = 500
Us = rnd.rand(N_units, 1)

ave_degree = 100
W_graph = nx.connected_watts_strogatz_graph(N_units, ave_degree, 0.5)

W = np.asarray(nx.to_numpy_matrix(W_graph))/(ave_degree) 

U_w = rnd.randn(N_units, N_units) * 10;
W = W * U_w

N_in = 1
OutU = 0
In = rnd.choice(range(1,N_units), size=N_in, replace=True)
In = [400]
```

```python
# alpha = 0.01

N_T = 1000
Ext_input = external_in(N_in, np.array(range(N_T)))
learning_T = 450

Warm_up_T = 50
record = np.zeros((N_units, N_T))
for t in range(N_T):  
    _preUs = W.dot(Us)

    Us = process(_preUs)
    if t < learning_T and t > Warm_up_T:
        Us[In] += Ext_input[:, [t]]  
    record[:, t] = np.ravel(Us)
```

```python
fig1, ax = subplots(ncols=2, figsize=(16, 8))
img1 = ax[0].matshow(W)
plt.colorbar(img1, ax=ax[0], orientation='horizontal')
print(In)

final_corr = np.abs(np.corrcoef(record[:,learning_T-40:learning_T+40]))
img2 = ax[1].matshow(final_corr)
plt.colorbar(img2, ax=ax[1], orientation='horizontal')

fig2, ax2 = subplots(figsize=(16, 8))
img = ax2.matshow(record)
plt.colorbar(img, ax=ax2, orientation='horizontal')

plt.figure(figsize=(10, 6))


plt.plot(Ext_input.T[learning_T-40:learning_T+40])
best_matches = []
for i in In:
    best_matches.extend(np.argsort(np.ravel(final_corr[i,:]))[-2:])
plt.plot(record[best_matches,learning_T-40:learning_T+40].T)

line_lbs = ['in']
line_lbs.extend([str(i) for i in best_matches])
plt.legend(line_lbs, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
print(In)
```
